# FNO for Darcy Flow Reconstruction with Sparse Observations: Mask Raw

In this notebook, we explore different methods for embedding sparse observation data to train a Fourier Neural Operator (FNO). The objective is to reconstruct the full pressure field from sparse sensor measurements.

## Problem Formulation
Given sparse observations at sensor locations $x_k$, our goal is to learn an operator:
$$
\mathcal{L}: \text{Embedding}(u_k, x_k) \mapsto u(x),
$$
where $u_k(x)$ is the observed field at the sensor locations and $x_k$ are the corresponding coordinates.

## Sparse Data Embedding Methods
To make the sparse input suitable for FNO, we test multiple embedding strategies:

1. Direct Sparse Field Embedding (Mask Raw)  
   Using the raw observed field:
   $$u_k(x) = M(x) u(x)$$
   where \( M(x) \) is a binary mask indicating sensor locations.

2. Mask Embedding
   We create an embedding from:
   - **raw observed field** $u_k(x)$
   - **X-coordinates** $D_x(x)$
   - **Y-coordinates** $D_y(x)$

   These are concatenated into:
   $$E(x) = [u_k(x), D_x(x), D_y(x)]$$
   and processed via a **1×1 convolution layer**:
   $$E'(x) = \text{Conv}_{1×1}(E(x))$$
   The learned embedding $E'(x)$ (size of (grid_x, grid_y, num_of_features (like 1 for this case))) is then used as input to FNO.

3. Voronoi Embedding:

## Operator Learning
In this notebook, we specifically use **mask embedding**:
$$
\mathcal{L}: E'(x) \mapsto u(x)
$$
where $E'(x)$ is the mask embedding obtained from $M, D_x, D_y$.

Future experiments will compare this approach with other embedding strategies to analyze their impact on the reconstruction accuracy.


```python
# packages
import torch
import numpy as np
import matplotlib.pyplot as plt
import sys
from neuralop.models import FNO
from neuralop import Trainer
from neuralop.training import AdamW
from neuralop.layers.embeddings import GridEmbedding2D
from neuralop.data.datasets import load_darcy_flow_small
from neuralop.utils import count_model_params
from neuralop import LpLoss, H1Loss

device = 'cpu'
```

```python
# load original dataset
train_loader, test_loaders, data_processor = load_darcy_flow_small(
        n_train=1000, batch_size=32,
        test_resolutions=[16], n_tests=[100],
        test_batch_sizes=[32],
)
data_processor = data_processor.to(device)
```

```python
# set sensor locations
positions = np.array([
    [0,15], [2, 15], [5, 15], [10, 15],
    [15,0], [15,2], [15,5], [15,6], [15,10], [15,13], [15,15],
    [5,5], [10,10], 
    [3,3], [3,8], [3,12], [8,3], [8,8], [8,12]
])

# positions
mask = np.zeros((1, 16, 16))
for pos in positions:
    mask[0, pos[0], pos[1]] = 1

plt.imshow(mask[0], origin='lower')
plt.show()
```

```python
# generate sparse dataset with original dataset
from torch.utils.data import DataLoader
mask = torch.tensor(mask, device=device)

# modify x to y*mask for train dataset
mod_train_data = []
for batch in train_loader.dataset:
    y = batch['y'].to(device).float()
    x_mod = (y * mask).float()
    mod_train_data.append({'x': x_mod, 'y': y})

mod_train_loader = DataLoader(mod_train_data, batch_size=32, shuffle=True)

# modify x to y*mask for test dataset
mod_test_loaders = {}
for res, test_loader in test_loaders.items():
    mod_test_data = []
    for batch in test_loader.dataset:
        y = batch['y'].to(device).float()
        x_mod = y * mask.float()
        mod_test_data.append({'x': x_mod, 'y': y})

    mod_test_loaders[res] = DataLoader(mod_test_data,
                                       batch_size=test_loader.batch_size,
                                       shuffle=False,
                                       num_workers=0,
                                       pin_memory=True)

# for batch in mod_test_loader:
#     print("x shape:", batch['x'].shape)
#     print("y shape:", batch['y'].shape)
#     break  # 只打印一个 batch
```

```python
# check modified dataset

index = 90
data = mod_train_loader.dataset[index]
x = data['x']
y = data['y']

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(2, 2, 1)
ax.annotate(f"Index {index}", xy=(-0.15, 0.5), 
             xycoords="axes fraction",
             ha="right", va="center", rotation=90)
masked_x = np.ma.masked_where(x[0] == 0, x[0])
mod_cmap = plt.cm.viridis.copy()
mod_cmap.set_bad(color='white') 
ax.imshow(masked_x, cmap=mod_cmap, origin='lower')
ax.set_title('input x: sparse field $u_k(x)$')
ax = fig.add_subplot(2, 2, 2)
ax.imshow(y.squeeze(), origin='lower')
ax.set_title('input y: full field $u(x)$')
plt.show()
```

```python
# model setup
model = FNO(n_modes=(16, 16),
             in_channels=1,
             out_channels=1,
             hidden_channels=64,
             positional_embedding="grid",
             projection_channel_ratio=2)
model = model.to(device)

n_params = count_model_params(model)
print(f'\nOur model has {n_params} parameters.')
sys.stdout.flush()
```

```python
# model setup
optimizer = AdamW(model.parameters(),
                                lr=4e-3,
                                weight_decay=1e-4)
# scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)

# loss functions
l2loss = LpLoss(d=2, p=2)
h1loss = H1Loss(d=2)

train_loss = h1loss
eval_losses={'h1': h1loss, 'l2': l2loss}


print('\n### MODEL ###\n', model)
print('\n### OPTIMIZER ###\n', optimizer)
print('\n### SCHEDULER ###\n', scheduler)
print('\n### LOSSES ###')
print(f'\n * Train: {train_loss}')
print(f'\n * Test: {eval_losses}')
sys.stdout.flush()
```

```python
# train
trainer = Trainer(model=model, n_epochs=100,
                  device=device,
                  data_processor=data_processor,
                  wandb_log=False,
                  eval_interval=1,
                  use_distributed=False,
                  verbose=True)

trainer.train(train_loader=mod_train_loader,
              test_loaders=mod_test_loaders,
              optimizer=optimizer,
              scheduler=scheduler,
              regularizer=False,
              training_loss=train_loss,
              eval_losses=eval_losses,
              save_every=10,
              save_dir="./ckpt/sparse_mask/",
              # resume_from_dir="./ckpt/sparse_mask/"
              )
```

```python
# plot train and eval loss
import re
import matplotlib.pyplot as plt

with open("log.txt", 'r') as f:
    log = f.read()

train_loss = [float(x) for x in re.findall(r"avg_loss=([0-9.]+)", log)]
eval_loss = [float(x) for x in re.findall(r"16_h1=([0-9.]+)", log)]

train_loss_rel = [x / train_loss[0] for x in train_loss]
eval_loss_rel = [x / eval_loss[0] for x in eval_loss]

# plot
fig, ax = plt.subplots()
ax.plot(train_loss_rel, label='Train')
ax.plot(eval_loss_rel, label='Eval')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.legend()
plt.show()
```
